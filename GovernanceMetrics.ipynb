```python
#
# Load Governance Information for the current project
# List all Governed Bundle for the current project
#

import requests
import os
import json
from collections import defaultdict

class DominoAuth(requests.auth.AuthBase):
    def __init__(self, base_url=os.environ["DOMINO_API_PROXY"]):
        self.base_url = base_url

    def __call__(self, r):
        res = requests.get(f"{self.base_url}/access-token")
        r.headers["Authorization"] = f"Bearer {res.text}"
        return r


host = "https://se-demo.domino.tech/"
project_id = os.environ.get("DOMINO_PROJECT_ID")

if not project_id:
    raise ValueError("DOMINO_PROJECT_ID environment variable is not set.")
    
# Get all deliverables (bundles) for the project
res = requests.get(
    f"{host}/guardrails/v1/deliverables?project_id={project_id}",
    auth=DominoAuth(),
)

# Parse and print response
bundles = res.json()
print(json.dumps(bundles, indent=4))

```

```python
#
# List All Bundles 
#

if bundles["data"]:
    print("Bundles found:")
    for bundle in bundles["data"]:
        print(f"Name: {bundle.get('name', 'Unnamed')}, Bundle ID: {bundle['id']}, Policy ID: {bundle.get('policyId')}, Stage: {bundle.get('stage')}, State: {bundle.get('state')}")
else:
    print("No bundles found.")
```

```python
#
# Group bundles by policy name
#

bundles_per_policy = defaultdict(list)
for bundle in bundles.get("data", []):  # Use the `bundles` variable
    policy_name = bundle.get("policyName", "No Policy Name")  # Extract policy name
    bundle_name = bundle.get("name", "Unnamed Bundle")  # Extract bundle name
    bundles_per_policy[policy_name].append(bundle_name)  # Group by policy name

# Print results
print("Bundles per Policy:")
for policy_name, bundle_names in bundles_per_policy.items():
    print(f"\nPolicy Name: {policy_name}")
    print(f"  Number of Bundles: {len(bundle_names)}")
    print("  Bundles:")
    for bundle_name in bundle_names:
        print(f"    - {bundle_name}")
```

```python
# Group bundles by policy name and state
bundles_per_policy_state = defaultdict(lambda: defaultdict(int))

for bundle in bundles.get("data", []):  # Use the `bundles` variable
    policy_name = bundle.get("policyName", "No Policy Name")  # Extract policy name
    bundle_state = bundle.get("state", "No State")  # Extract state
    bundles_per_policy_state[policy_name][bundle_state] += 1  # Increment count

# Print results
print("Bundles per Policy and State:")
for policy_name, states in bundles_per_policy_state.items():
    print(f"\nPolicy Name: {policy_name}")
    for state, count in states.items():
        print(f"  State: {state}, Number of Bundles: {count}")
```

```python
# Group bundles by policy name and stage, and store detailed information
bundles_per_policy_stage = defaultdict(lambda: defaultdict(list))

for bundle in bundles.get("data", []):  # Use the `bundles` variable
    policy_name = bundle.get("policyName", "No Policy Name")  # Extract policy name
    bundle_stage = bundle.get("stage", "No Stage")  # Extract stage
    bundle_details = {
        "name": bundle.get("name", "Unnamed Bundle"),
        "createdAt": bundle.get("createdAt", "Unknown Date"),
        "state": bundle.get("state", "Unknown State"),
    }
    bundles_per_policy_stage[policy_name][bundle_stage].append(bundle_details)

# Print results
print("Bundles per Policy and Stage with Details:")
for policy_name, stages in bundles_per_policy_stage.items():
    print(f"\nPolicy Name: {policy_name}")
    for stage, bundles in stages.items():
        print(f"  Stage: {stage}, Number of Bundles: {len(bundles)}")
        for bundle in bundles:
            print(f"    - Bundle Name: {bundle['name']}")
            print(f"      Created At: {bundle['createdAt']}")
            print(f"      State: {bundle['state']}")
```
