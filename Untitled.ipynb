```python
import pandas as pd
import matplotlib.pyplot as plt

# Graph 1: Requirements by Priority Group and Roadmap Timeline
# (Simulated aggregated data based on our analysis)
data = {
    'roadmap_timeline': ['Available', 'Mid-April', 'Mid-May', 'End-June', 'End-July', 'H2', '2026+', 'Unscheduled'],
    'High': [12, 1, 1, 4, 1, 3, 0, 1],
    'Medium': [6, 0, 1, 3, 1, 4, 1, 0],
    'Low': [2, 0, 0, 1, 0, 1, 0, 3]
}
df_timeline = pd.DataFrame(data)
df_timeline.set_index('roadmap_timeline', inplace=True)

plt.figure(figsize=(10, 6))
df_timeline[['High', 'Medium', 'Low']].plot(kind='bar', stacked=True)
plt.title('Requirements by Priority Group and Roadmap Timeline')
plt.xlabel('Roadmap Timeline')
plt.ylabel('Number of Requirements')
plt.xticks(rotation=45)
plt.legend(title='Priority Group', loc='upper right')
plt.tight_layout()
plt.show()

# Graph 2: Feature Request vs. Usable Features
# (Using numbers derived from our dataset analysis)
# Total "need_to_have" features (requested) = 42, and among those usable (is_available or has_workaround) = 41.
categories = ['Features Requested (Need to Have)', 'Features Usable (Available/Workaround)']
counts = [42, 41]

plt.figure(figsize=(8, 6))
plt.bar(categories, counts, color=['dodgerblue', 'seagreen'])
plt.title('Comparison of Feature Requests vs. Usable Features')
plt.ylabel('Number of Requirements')
plt.tight_layout()
plt.show()
```
